```python

labeled_pct = 0.001
save_PATH = '/data/ddmg/frank/shuvom/tutorials/civilcomments_data/labeled_{}/'.format(labeled_pct)

# true confusions
confusion = np.load(save_PATH + 'confusions_{}_sampled.npy'.format(0))

# print(confusion)
theta = np.load(save_PATH + 'theta_{}_sampled.npy'.format(0))

markers = ["o", "x", "v"]
for i in range(3):
    # print()
    plt.scatter([theta[i,0,0], theta[i,1,1]],
                [confusion[i,0,0], confusion[i,1,1]], c="blue", marker=markers[i], label="on-diagonal")
    plt.scatter([theta[i,0,1], theta[i,1,0]],
            [confusion[i,0,1], confusion[i,1,0]], c="green", marker=markers[i], label="off-diagonal")
#     plt.scatter(np.concatenate((theta[i,0,0], theta[i,1,1])),
#             np.concatenate((confusion[i,0,0], confusion[i,1,1])), c="blue", marker=markers[i], label="on-diagonal")
#     plt.scatter(np.concatenate((theta[i,0,1], theta[i,1,0])),
#             np.concatenate((confusion[i,0,1], confusion[i,1,0])), c="green", marker=markers[i], label="off-diagonal")
plt.plot([0,1], [0,1], c="red")
plt.xlabel("Estimated Confusion Matrix")
plt.ylabel("True Confusion Matrix")
plt.title("D-S Confusion Matrix Estimation (civilcomments), labeled_pct = {}".format(labeled_pct))  
plt.legend()
# theta[:,1,1]
```
